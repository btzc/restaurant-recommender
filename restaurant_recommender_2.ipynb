```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_cuisine = pd.read_csv('./inputs/chefmozcuisine.csv')
df_cuisine.head()
```

```python
df_parking = pd.read_csv('./inputs/chefmozparking.csv')
df_parking.head()
```

```python
df_geoplaces = pd.read_csv('./inputs/geoplaces2.csv')
df_geoplaces.head()
```

```python
df_ratings = pd.read_csv('./inputs/rating_final.csv')
df_ratings.head()
```

```python
df = pd.merge(df_ratings, df_geoplaces, on='placeID')
df.head()
```

```python
df['overall_rating'] = (df['rating'] + df['food_rating'] + df['service_rating'])/3
df.head()
```

```python
%matplotlib inline
```

```python
df.groupby('name')['overall_rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('name')['overall_rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('name')['overall_rating'].mean())
ratings.head()
```

```python
ratings['number of ratings'] = pd.DataFrame(df.groupby('name')['overall_rating'].count())
ratings
ratings.head()
```

```python
ratings['number of ratings'].hist(bins=50)
```

```python
ratings['overall_rating'].hist(bins=50)
```

```python
sns.jointplot(x='overall_rating',y='number of ratings',data=ratings,alpha=0.5)
```

```python
restmatrix = df.pivot_table(index='userID',columns='name',values='overall_rating')
restmatrix.head()
```

```python
ratings.sort_values('number of ratings', ascending=False).head(10)
```

```python
oriental_user_ratings = restmatrix['Restaurant Oriental Express']
posada_user_ratings = restmatrix['La Posada del Virrey']
```

```python
oriental_user_ratings.head()
```

```python
sim_to_oriental = restmatrix.corrwith(oriental_user_ratings)
sim_to_oriental.head()
```

```python
sim_to_posada = restmatrix.corrwith(posada_user_ratings)
sim_to_posada.head()
```

```python
corr_oriental = pd.DataFrame(sim_to_oriental,columns=['Correlation'])
corr_oriental.dropna(inplace=True)
corr_oriental.sort_values('Correlation',ascending=False).head()
```

```python
corr_oriental = corr_oriental.join(ratings['number of ratings'])
corr_oriental.head()
```

```python
corr_oriental[corr_oriental['number of ratings']>15].sort_values('Correlation',ascending=False).head(10)
```

```python
corr_posada = pd.DataFrame(sim_to_posada,columns=['Correlation'])
corr_posada.dropna(inplace=True)
corr_posada.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_posada = corr_posada.join(ratings['number of ratings'])
corr_posada.head()
```

```python
corr_posada[corr_posada['number of ratings']>15].sort_values('Correlation',ascending=False).head(10)
```
