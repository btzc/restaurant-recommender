## 1.0 Introduction
Restaurants can be categorized based on many different features. The primary source for people to decide which new restaurant to try out heavily depends on ratings left by other users. However, these ratings do not always tell the full story, as different users can have very different preferences.

Word of mouth recommendations are one way people can decide where to go next based on how they view the person they are getting this recommendation from. If they believe said person has a good sense of taste, or one similar to themself, they are more likely to go to wherever they recommend. But a problem comes in where maybe a user does not have a large community around them, or believes none of the people around them relate closely enough to themselves. 

This is where online rating services come in. These services can build out entire matricies of user preferences and rating information, and use it to predict what this user may enjoy.

Recommender systems are algorithms who aim to suggest relevent items to users based on previous information. These systems are critical to companies for efficiently being able to generate income to stand out from competitors (https://towardsdatascience.com/introduction-to-recommender-systems-6c66cf15ada). An example of how important recommender systems are is the Netflix challenge, where they offered a grand prize of $1,000,000 USD to anyone who could improve their collaboriate filtering algorithm (https://en.wikipedia.org/wiki/Netflix_Prize).

This notebook will compare and contrast two recommender algorithms: Content-Based and Collaborative Filtering. It will also discuss the trimming of features to avoid overfitting and reduce computation time, while not sacrificing accuracy. 

First, we cover data preprocessing where we prepare the information gathered into the proper structure required. Next, we discuss our implementation of both Content-Based and Collaborative Filtering algorithms. Finally, we evaluate our solutions using the Root Mean Square Error (RMSE) method and compare the different methods above.

## 3.0 Implementation of Models
This section discusses the implementation choices for each model. The two models implemented in this notebook were content-based and collaborative filtering. For collaborative filtering, this notebook only explores memory based and not model based. The following is a generalization of our implementation approach:
1. Create a matrix where the indexes are userID's and the columns are the names of restaurants.
2. Replace any missing data
3. Calculate the pairwise distances using cosine similarity given the training matrix
4. Make predictions given the output of step 3, and also the training matrix.

Pairwise distances allow us to correlate the vector array create in step one. It works by computing the distances given a metric, such as cosine similarity, and then returns a distance matrix (https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise_distances.html). The values in the distance matrix represent how correlated each item is. The larger the number, the less correlated the item. Pairwise distances are computed by calculating `1-metric`. So in this case, our pairwise distances are calculated by computing `1-cosine_sim(matrix)`

### 3.1 Content-Based Model Implementation
Content-based recommendation systems work by taking into account the data a user has already provided. This could be anything from clicking a link, to providing a rating (https://www.analyticsvidhya.com/blog/2015/08/beginners-guide-learn-content-based-recommender-systems/). The more inputs the system gathers about the user, the more accurate it becomes. The below is our implementation of a content-based recommender system only taking into account the `overall_rating`.

````python
# imports
from sklearn.metrics.pairwise import pairwise_distances
````

````python
# put the train and split data into pivot tables
train_data_pivot = train_data.pivot_table(index='userID',columns='name',values='overall_rating')
test_data_pivot = test_data.pivot_table(index='userID',columns='name',values='overall_rating')

# fill any missing data with 0's
train_data_pivot.fillna(0, inplace=True)
test_data_pivot.fillna(0, inplace=True)
````

````python
# create the train matrix
train_data_mat = train_data_pivot.values
train_data_mat
````

````python
# create the test matrix
test_data_mat = test_data_pivot.values
test_data_mat
````

````python
# compute the pairwise distance with the `cosine` metric given the training matrix
content_rating_similarity = pairwise_distances(train_data_mat.T, metric='cosine')
content_rating_similarity
````

### 3.2 Collaborative-Filtering Model Implementation
Collaborative-filtering (CF) works off of other users and not just the user themselves. This algorithm works mostly by presuming that is two users have previously purchased something similar, those users are likely to purchase a product that the other has bought before (https://towardsdatascience.com/recommendation-systems-models-and-evaluation-84944a84fb8e). Compared to content-based, there are no feature corresponding to users or items. This means a rather large dataset is required for this method to be useful. In general, there are two types of CF implementations: memory-based and model-based. As mentioned above, we have only implemented memory-based CF models.

#### 3.2.1 Model Based Collaborative-Filtering
Model based collaborative filtering is when the utility matrix is memorized and any recommendations for users are made by querying against it. There are two types of model based collaborative-filtering methods: user-item and item-item. The user-item filtering method takes a particular user, finds users that are similar to that user based on ratings, and then recommends an item that those similar users have rated highly before. Comparatively, the item-item filtering method takes an item, finds users who have rated that item highly before, and finds other items that those users or similar users also rated highly. For our implementation of both these models, we used a single utiltiy matrix but just called the transpose of it when doing item-item filtering.

````python
# imports
import numpy as np
````

````python
# get number of unique users and restaurants to shape utility matrix
n_users = df.userID.nunique()
n_items = df.placeID.nunique()

print('Num. of Users: '+ str(n_users))
print('Num of Restaurants: '+str(n_items))
````

````python
#Create two user-item matrices, one for training and another for testing
train_data_cf_matrix = np.zeros((n_users, n_items))
for line in train_data.itertuples():
    train_data_matrix[line[1], line[2]] = line[3]  

test_data_cf_matrix = np.zeros((n_users, n_items))
for line in test_data.itertuples():
    test_data_cf_matrix[line[1], line[2]] = line[3]
````

````python
item_similarity = pairwise_distances(train_data_matrix.T, metric='cosine')
user_similarity = pairwise_distances(train_data_matrix, metric='cosine')
````

## 4.0 Evaluation
This section will discuss how we evaluated the different recommender models implemented. There were two steps to evaluating our models:
1. Predictions
2. Validation

Predictions are what the system uses to recommend the user the next item. Validation is how we know how well the system is executing.

### 4.1 Predictions
Being able to effectively predict what a user will rate an item they have not previously rated is the entire purpose of recommendation systems. The more accurately the system can predict the rating, the better the curated list of recommendations can be. For collaborative filtering, predictions are calculated in two ways: one for user-item, and the other for item-item.

#### 4.1.1 User-Item Prediction Calculation
The user-item prediction calculation is computed by getting the weighted average fo the deviations from similar users, and then adding these to the active user's mean rating. Deviations of ratings are used to account for any user associated biases (https://towardsdatascience.com/collaborative-filtering-based-recommendation-systems-exemplified-ecbffe1c20b1).

#### 4.1.2 Item-Item Prediction Calculation
The item-item prediction calculation is a little simpler than the user-item prediction calculation outlined above. This approach simply cares about the similarities between pairs of items. So, the rating for a target item is predicted by using weighted averages.

````python
# code from: https://www.udemy.com/course/python-for-data-science-and-machine-learning-bootcamp
# predict function
def predict(ratings, similarity, type='user'):
    # user-item predictions
    if type == 'user':
        mean_user_rating = ratings.mean(axis=1)
        ratings_diff = (ratings - mean_user_rating[:, np.newaxis]) 
        pred = mean_user_rating[:, np.newaxis] + similarity.dot(ratings_diff) / np.array([np.abs(similarity).sum(axis=1)]).T
    # item-item or content-based predictions
    elif type == 'item':
        pred = ratings.dot(similarity) / np.array([np.abs(similarity).sum(axis=1)])     
    return pred
````

````python
# generate predictions for three models

# cf item-item
cf_item_prediction = predict(train_data_matrix, item_similarity, type='item')

# cf user-item
cf_user_prediction = predict(train_data_matrix, user_similarity, type='user')

# cb ratings
cb_rating_prediction = predict(train_data_mat, content_rating_similarity, type='item')
cb_rating_prediction
````

### 4.2 Validation
One of the most difficult and important parts to recommendation systems is being able to validate the results returned. 

There are many different ways to validate results including recall, precision, mean absoulte error (MAE), mean squared error (MSE), and root mean squared error (RMSE). This notebook will focus on evaluating the models using RMSE (https://towardsdatascience.com/recommendation-systems-models-and-evaluation-84944a84fb8e).

````python
# imports
from sklearn.metrics import mean_squared_error
from math import sqrt
````

````python
# rmse calculation 
def rmse(prediction, test):
    prediction = prediction[test.nonzero()].flatten() 
    test = test[test.nonzero()].flatten()
    return sqrt(mean_squared_error(prediction, test))
````

````python
print('User-based CF RMSE: ' + str(rmse(cf_user_prediction, test_data_cf_matrix)))
print('Item-based CF RMSE: ' + str(rmse(cf_item_prediction, test_data_cf_matrix)))
print('Rating-only CB RMSE: ' + str(rmse(cb_rating_prediction, test_data_mat)))
````
