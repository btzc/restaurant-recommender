```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df_cuisine = pd.read_csv('./inputs/chefmozcuisine.csv')
df_cuisine.head()
```

```python
df_parking = pd.read_csv('./inputs/chefmozparking.csv')
df_parking.head()
```

```python
df_geoplaces = pd.read_csv('./inputs/geoplaces2.csv')
df_geoplaces.head()
```

```python
df_ratings = pd.read_csv('./inputs/rating_final.csv')
df_ratings.head()
```

```python
df = pd.merge(df_cuisine, df_geoplaces, on='placeID')
df = pd.merge(df, df_parking, on='placeID')
df.head()
```

```python
df['combined_features'] = df['Rcuisine']+' '+df['parking_lot']+' '+df['latitude'].map(str)+' '+df['longitude'].map(str)
df.head()
```

```python
df_combined_features = df[['placeID', 'name', 'combined_features']]
```

```python
cv = CountVectorizer()

count_matrix = cv.fit_transform(df_combined_features['combined_features'])
print(count_matrix)
```

```python
cosine_sim = cosine_similarity(count_matrix) 
cosine_sim
```

```python
oriental = 'Restaurant Oriental Express'
posada = 'La Posada del Virrey'

similar_rests_oriental = list(enumerate(cosine_sim[df[df.name == oriental].index[0]]))
df_similar_oriental = pd.DataFrame(similar_rests_oriental)
df_similar_oriental.columns = ['restaurant_index', 'cosine_similarity']
df_similar_oriental.head()
```

```python
df_similar_oriental = df_similar_oriental.join(df_combined_features)
df_similar_oriental
df_similar_oriental.sort_values('cosine_similarity',ascending=False).head(10)
```

```python
similar_rests_posada = list(enumerate(cosine_sim[df[df.name == posada].index[0]]))
df_similar_posada = pd.DataFrame(similar_rests_posada)
df_similar_posada.columns = ['restaurant_index', 'cosine_similarity']
df_similar_posada.head()
```

```python
df_similar_posada = df_similar_posada.join(df_combined_features)
df_similar_posada
df_similar_posada.sort_values('cosine_similarity',ascending=False).head(10)
```
